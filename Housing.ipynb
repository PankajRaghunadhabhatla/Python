```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing = pd.read_csv('housing.csv')
housing
```

```python
#. How many houses have a waterfront?
print(housing['waterfront'].sum())
```

```python
# How many houses have 2 floors?
two_floors = housing[housing['floors']==2]
two_floors.shape[0]
```

```python
waterfront_count = housing[housing['waterfront']==1]
waterfront_count.shape[0]
```

```python
#How many houses built before 1960 have a waterfront?
before_1960 = housing[(housing['yr_built']<1960) & (housing['waterfront']==1)]
before_1960.shape[0]
```

```python
#What is the price of the most expensive house having more than 4 bathrooms?
greater_than_4_bathrooms = housing[housing['bathrooms']>4]
greater_than_4_bathrooms.shape[0]
greater_than_4_bathrooms['price'].max()
```

```python
for i in housing.columns:
  if (housing[i].dtypes == 'int64' or housing[i].dtypes=='float64'):
    plt.hist(housing[i])
    plt.xlabel(i)
    plt.ylabel('count')
    plt.show()
```

```python
for i in housing.columns:
  if (housing[i].dtypes == 'int64' or housing[i].dtypes=='float64'):
    plt.boxplot(housing[i])
    plt.xlabel(i)
    plt.ylabel('count')
    plt.show()
```
