```python
X = df.drop(columns=["medv"]) #input columns
y = df["medv"] #target
```

    Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
```

```python
model = Sequential() # Empty model without any Hidden Layers
model.add(Input(13,))
model.add(Dense(units=2, activation= "relu"))
model.add(Dense(units=1))
```

```python
model.compile(optimizer="adam", loss="mean_squared_error") #MSE because of a regression problem or continuous output
```

```python
model.fit(x=X_train, y=y_train, epochs=150, validation_data=(X_test, y_test))
```

```python
model.history.history.keys()
```

```python
import matplotlib.pyplot as plt 
```

```python
train_loss = model.history.history["loss"]
val_loss = model.history.history["val_loss"]
```

```python
plt.plot(train_loss)
plt.plot(val_loss)
plt.xlabel("epochs")
plt.ylabel("MSE")
plt.legend(['Train', 'Test'])
plt.grid()
plt.show()
```
